## 线性方程及方程组
我们常常需要一个有唯一解的线性方程系统, 需要保证系数矩阵满秩full rank

由于计算机的浮点数是近似的，因此还需要保证矩阵的条件cond接近1

$$2x_{1}+3x_{2}=4 \\ 5x_{1}+4x_{2}=3$$

```python


# sympy
A = sympy.Matrix([[2,3],[5,4]])
b = sympy.Matrix([4,3])
A.rank()
sympy.N(A.condition_number())
A.norm()



# 解法1： 直接求解，逆矩阵
# 解法2： LU分解 A=LU （推荐）

A.LUdecomposition()
A.LUsolve(b)
```

```python
# scipy 近似解
A = np.array([[2,3],[5,4]])
b = np.array([4,3])

np.linalg.matrix_rank(A)
np.linalg.cond(A)
np.linalg.norm(A)
la.lu(A)
la.solve(A,b)
```

## 线性方程组的符号求解和数值求解的区别

求解如下线性方程组：

$$
\begin{pmatrix}
1 & \sqrt{p} \\
1 & \frac{1}{\sqrt{p}}
\end{pmatrix} 
\begin{pmatrix}
x_{1} \\
x_{2}
\end{pmatrix}
= 
\begin{pmatrix}
1 \\
2
\end{pmatrix}
$$



```python
import sympy as sp
sp.init_printing()

p = sp.symbols("p", positive=True)
A = sp.Matrix([[1, sp.sqrt(p)],[1,1/sp.sqrt(p)]])
b = sp.Matrix([1,2])

A
b
x_sym_sol = A.solve(b)
Acond = A.condition_number().simplify()
Acond
x_sym_sol[1].subs(p,0.8)
```

```python
import numpy as np

AA = lambda p: np.array([[1,np.sqrt(p)],[1,1/np.sqrt(p)]])
bb = np.array([1,2])

x_num_sol = lambda p: np.linalg.solve(AA(p),bb)

```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
#mpl.rcParams['text.usetex']=True
fig, axes = plt.subplots(1,2,figsize=(12,4))

p_vec = np.linspace(0.9,1.1,200)
colors = ['b', 'r']
labels = [r'x_{1}', r'x_{2}']

for n in range(2):
  x_sym = np.array([x_sym_sol[n].subs(p,pp) for pp in p_vec])
  x_num = np.array([x_num_sol(pp)[n] for pp in p_vec])
  axes[0].plot(p_vec, (x_num-x_sym)/x_sym, colors[n], label=labels[n])
  axes[0].legend()
axes[1].plot(p_vec,[Acond.subs(p,pp) for pp in p_vec])
```

## 二分法求解非线性方程 

该方法的适用于连续函数$f(x)$，该方法的步骤如下：

- 首先设置一个$x$的区间$[a,b]$，保证$f(a)f(b)<0$，根据零点定理，$f(x)$在$[a,b]$中至少存在一个零点，即至少存在一个关于$f(x)=0$的解;
- 根据零点存在定理，只要我们可以不断压缩这个区间，让解的范围足够小，而二分法的压缩方法就是对区间进行对半分，直到找到符合误差范围时，则选择$(a+b)/2$为我们的数值解


```python
import numpy as np
import matplotlib.pyplot as plt
f = lambda x: np.exp(x) -2 
tol = 0.1
a, b = -2, 2
x = np.linspace(-2.1,2.1,1000)

fig, ax = plt.subplots(figsize=(12,4))

ax.plot(x, f(x))
ax.axvline(0, ls=':', color='k')

ax.set_xticks(np.arange(-2,3,1))

fa, fb = f(a), f(b)

ax.plot(a,fa,'ko')
ax.plot(b,fb,'ko')
ax.text(a,fa+0.5,r"a")
ax.text(b,fb+0.5,r"b")

n = 1
while b-a >tol:
  m = a + (b-a)/2
  fm = f(m)
  
  ax.plot(m,fm, 'ko')
  ax.text(m,fm-0.5,f"m{n}")
  n +=1
  
  if np.sign(fa) == np.sign(fm):
    a, fa = m, fm
  else:
    b, fb = m, fm
    
ax.plot(m,fm, 'r*')
ax.annotate(f"Root approximately at {m:.3f}", xy=(m,fm),xytext=(-150,50),textcoords='offset points', arrowprops=dict(arrowstyle="->"))
print(f(m))
```

## 牛顿法 (比二分法更快收敛)
如果一个函数$f(x)$是二次连续可导，且在[a,b]范围内有零点且一阶导不为0,则可以使用牛顿法来进行$f(x)=0$求解。

牛顿法的求解逻辑是使用一阶泰勒近似，也就是切线方程的解来获得近似解：

$$
f(x) = f(x_{0})+f'(x_{0})(x-x_{0})=0
$$
即获得：
$$
x = x_{0}-\frac{f(x_{0})}{f'(x_{0})}
$$

也就是使用迭代式：
$$
x_{n+1} = x_{n} -\frac{f(x_{n})}{f'(x_{n})}
$$


```python
# 求e^x-2=0
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
s_x = sp.symbols("x")
s_f = sp.exp(s_x)-2 # 原函数

f = sp.lambdify(s_x, sp.exp(s_x)-2, 'numpy')
fp = sp.lambdify(s_x, sp.diff(s_f,s_x), 'numpy')

x = np.linspace(-2.1,2.1,1000)

fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,f(x))
ax.axhline(0,ls=':')

n = 0
tol = 0.1
x0 = 2

while f(x0) > tol:
  x1 = x0 - f(x0)/fp(x0)
  ax.plot([x0, x0],[0,f(x0)],'k:')
  ax.plot(x0, f(x0),'ko')
  ax.plot([x0,x1],[f(x0),0],'k')
  x0 = x1

ax.plot(x1,f(x1),'r*')
ax.annotate(f"Root approximately at {x1:.3f}",
            xy=(x1,f(x1)),
            xytext=(-150,+50),
            textcoords='offset points',
            arrowprops=dict(arrowstyle="->"))

print(f(x1))
```

## 割线法
牛顿法的应用必须要求$f'(x)$的显式表达式，而且要求在近似点的导数不能为0。这就限制了很多函数，比如：

- 没法写出显式的导数表达式
- 很难保证在所有切线方程的零点对应的导数为0

为了解决第一个问题，可以使用如下式子来近似计算导数：

$$
f'(x) = \frac{f(x_{1})-f(x_{0})}{x_{1}-x_{0}}
$$
这时候牛顿法的迭代式子变为：

$$
x_{2} = x_{1}-\frac{f(x_{1})(x_{1}-x_{0})}{ f(x_{1})-f(x_{0})}
$$

这意味着我们需要设定两个初始值，而且两个初值的间隔要足够小用于近似导数。其他的就和牛顿法一样了。这样能保证不需要计算导数就可以进行牛顿法类似的迭代

```python
# 求e^x-2=0
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
s_x = sp.symbols("x")
s_f = sp.exp(s_x)-2 # 原函数

f = sp.lambdify(s_x, sp.exp(s_x)-2, 'numpy')
fp = sp.lambdify(s_x, sp.diff(s_f,s_x), 'numpy')

x = np.linspace(-2.1,2.1,1000)

fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x,f(x))
ax.axhline(0,ls=':')


tol = 0.1
x0 = 2
xn1 = 1.9

while f(x0) > tol:
  x1 = x0 - f(x0)*(x0-xn1)/(f(x0)-f(xn1))
  ax.plot([x0, x0],[0,f(x0)],'k:')
  ax.plot(x0, f(x0),'ko')
  ax.plot([x0,x1],[f(x0),0],'k')
  x0, xn1 = x1, x0

ax.plot(x1,f(x1),'r*')
ax.annotate(f"Root approximately at {x1:.3f}",
            xy=(x1,f(x1)),
            xytext=(-150,+50),
            textcoords='offset points',
            arrowprops=dict(arrowstyle="->"))

print(f(x1))
```

```python
# Scipy的optimize模块提供了现成的求解方法
# 注意：与我们自己迭代的程序不同，这些现成程序没法设置函数值与0的距离来终止循环，而是两次迭代根的距离来制止，也就是二分法的终止条件
from scipy.optimize import root_scalar

x = np.linspace(-2.1,2.1,1000)
f = lambda x: np.exp(x)-2
fp = lambda x: np.exp(x)
# 牛顿法
sol = root_scalar(f,method='newton',x0=2,xtol=0.1,fprime=fp)
print(sol.root)
# 割线法
sol = root_scalar(f,method='secant',x0=1.9,x1=2,xtol=0.1)
print(sol.root)
# 二分法
sol = root_scalar(f,method='bisect',bracket=[-2,2],xtol=0.1)
print(sol.root)
# 二分法改进
```
